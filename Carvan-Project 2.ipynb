```python
import numpy as np
import pandas as pd
```

```python
train= pd.read_csv("D:/Python/Edvancer/Project 2/carvan_train (1).csv")
test=pd.read_csv("D:/Python/Edvancer/Project 2/carvan_test (1).csv")
```

```python
train.shape
```

```python
train.dtypes
```

```python
train["V86"].value_counts(normalize=True,dropna=False)
```

```python
test["V86"]=0
train["Temp"]="train"
test["Temp"]="test"
combi=pd.concat([train,test],axis=0)
```

```python
combi.apply(lambda x: sum(x.isnull()))
```

```python
combi.describe(percentiles=[0.01,0.10,0.90,0.95,0.99]).T
```

```python
combi["V1"]=combi["V1"].astype('str')
```

```python
combi["V5"]=combi["V5"].astype('str')
```

```python
dummy=pd.get_dummies(combi[["V1","V5"]],drop_first=False)
```

```python
combi_updated=pd.concat([combi,dummy],axis=1)
```

```python
combi_latest=combi_updated.drop(["V1","V5"],axis=1)
```

```python
training=combi_latest.loc[combi["Temp"]=="train"]
```

```python
training.drop(["Temp"],axis=1,inplace=True)
```

```python
testing=combi_latest.loc[combi["Temp"]=="test"]
```

```python
x_train=training.drop(["V86"],axis=1)
```

```python
y_train=training["V86"]
```

```python
testing.drop(["V86","Temp"],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(324)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x_train,y_train,test_size=0.25,stratify=y_train)
```

```python
from xgboost.sklearn import XGBClassifier
```

```python
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
```

```python
param_dist = {
              "max_depth": [2,3,4,5,6,8],
              "learning_rate":[0.01,0.05,0.1,0.3,0.5],
    "min_child_weight":[4,5,6],
              "subsample":[i/10.0 for i in range(6,10)],
 "colsample_bytree":[i/10.0 for i in range(6,10)],
               "reg_alpha":[1e-5, 1e-2, 0.1, 1, 100],
              "gamma":[i/10.0 for i in range(0,5)],
    "n_estimators":[100,500,700,1000],
    'scale_pos_weight':[2,3,4,5,6,7,8,9]
    
              }
```

```python
clf=XGBClassifier(objective='binary:logistic')
```

```python
n_iter=10

random_search=RandomizedSearchCV(clf,n_jobs=-1,verbose=2,cv=10,n_iter=n_iter,scoring='roc_auc',
                                 param_distributions=param_dist)
```

```python
random_search.fit(X_train,Y_train)
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.5f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```python
report(random_search.cv_results_,5)
```

```python
xgb_best= XGBClassifier(objective='binary:logistic',subsample= 0.9, scale_pos_weight=3, reg_alpha=0.01,n_estimators=500,min_child_weight=5,max_depth=2,learning_rate=0.01,gamma=0.0,colsample_bytree=0.8)
```

```python
xgb_best.fit(X_train,Y_train)
```

```python
p=xgb_best.predict_proba(X_test)[:,1]
```

```python
p
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, accuracy_score, fbeta_score, precision_recall_curve
```

```python
roc_auc_score(Y_test,p)
```

```python
precision, recall, _ = precision_recall_curve(Y_test,p)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.plot(recall, precision, marker='.', label='xgboost')
plt.xlabel('Recall')
plt.ylabel('precision')
# show the legend
plt.legend()
# show the plot
plt.show()
```
